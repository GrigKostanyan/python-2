## Problem 1

```
Let say you want to send a message to someone. You have 2 options - you can Email the message or SMS the message to the corresponding person. So, you have two options to send the message and the client side code will use one of the implementations to send the message to the corresponding person. Use Bridge Design Pattern to implement the logic with classes of your choice and make sure to test the implementation with some concrete objects.
```

````python
class EmailSender:
    def send(self, message):
        print("Email: " + message)

class SMSSender:
    def send(self, message):
        print("SMS: " + message)
        
        
class Client:
    def __init__(self, Sender):
        self.sender = Sender()
    def sendMessage(self, message):
        self.sender.send(message)
        
e = Client(EmailSender)
e.sendMessage('Hi')

s = Client(SMSSender)
s.sendMessage('Hi')
````

## Problem 2

```
Follow the diagram below and use Composite Design Pattern to implement and test the logic. Add any other classes of your choice. The top of your hierchy will be the box which will contain Instrument Collection which will contain instruments. 
```

````python
class InstrumentCollection:
    pass

class Pen(InstrumentCollection):
    def Make(self):
        print('Making pen...')

class Pencil(InstrumentCollection):
    def Make(self):
        print('Making pencil...')

class Rubber(InstrumentCollection):
    def Make(self):
        print('Making rubber...')

class Box(InstrumentCollection):
    def __init__(self):
        self.col = []
    
    def Add(self, instrumentCollection):
        self.col.append(instrumentCollection)
        
    def Remove(self, instrumentCollection):
        self.col.remove(instrumentCollection)
        
    def GetChild(self, n):
        return self.col[n]
    
    def Make(self):
        for i in self.col:
            i.Make()
            
pen = Pen()
pencil = Pencil()
rubber = Rubber()

b = Box()

b.Add(pen)
b.Add(pencil)
b.Add(rubber)

b2 = Box()

b2.Add(pen)
b2.Add(pencil)
b2.Add(rubber)

b.Add(b2)

b.Make()
````
